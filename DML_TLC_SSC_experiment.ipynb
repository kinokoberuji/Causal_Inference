## Cấu trúc mô hình Causal inference

Ta thiết lập một cấu trúc mô hình như sau:

Block can thiệp điều trị T_col là một array 2D, tương ứng với 2 biến nhị phân là 'Vasdilator' và 'ImSupres', tên 2 loại thuốc cần được đánh giá hiệu quả.

Outcome Y là biến TLC_z, hay Z_score của dung tích phổi, là một biến liên tục, giả định có phân phối chuẩn với vị trí trung tâm = 0 và SD = 1

Block X_col là các Covariate và/hoặc Counfounders (Ta biết chắc chắn là chúng có ảnh hưởng lên Outcome và/hoặc xác suất được điều trị bằng 1 đến 2 thứ thuốc (Trên thực tế, thuốc ức chế miễn dịch được chỉ định cho các bệnh nhân có triệu chứng viêm và xơ hóa nặng, có xơ phổi, suy giảm chức năng hô hấp...), thuốc dãn mạch được chỉ định cho bệnh nhân có Tăng áp phổi (PAH); và dĩ nhiên các chỉ số chức năng như FEV1, FVC, TLCO, KCO, VA... đều liên hệ với TLC (covariates).

Duration là thời gian từ khi chẩn đoán đến ngày làm xét nghiệm, tính bằng năm, đây là 1 covariate quan trọng vì nó sẽ được dùng để đánh giá tính ổn định của hiệu quả điều trị theo thời gian.

Block W bao gồm những biến vô thưởng vô phạt, chúng không có vai trò rõ ràng đối với kết quả điều trị lẫn xác suất được điều trị.

Dĩ nhiên ta loại bỏ tất cả quan sát có missing value

```python
T_col = ['Vasdilator','ImSupres',]

Y_col = ['TLC_z',]

W_col = ['Cur_Smoke','Ex_Smoke','Gender',
         'L','D','PreSc','CRP','Hb',]

X_col = ['Duration',
         'Lung_fib','PAH_Pre','PAH_Post',
         'FEV1_z','FVC_z','TLCO_z','KCO_z','VA_z']

tlc_df = df[T_col + X_col + Y_col + W_col].dropna(how='any')

T = tlc_df[T_col]
Y = tlc_df[Y_col]
X = tlc_df[X_col]
W = tlc_df[W_col]
```

```python
X.shape
```

Sau khi loại bỏ missing value, còn lại 574 quan sát trong dữ liệu X,Y,W và T

## Thăm dò trực quan

Ta dựng một số biểu đồ để thăm dò mối liên hệ giữa các covariates và outcome:

```python
target = 'TLC_z'
x_vars = ['Duration','FEV1_z','FVC_z','VA_z']

fig, axs = plt.subplots(ncols=len(x_vars), figsize=(22,5))

for i,v in enumerate(x_vars):
    g = sns.regplot(x=X[v].values, 
                y=Y[target].values,
                lowess = True, marker = ".",
                scatter_kws={'alpha':0.3, 'color':'grey'},
                ax = axs[i])
    g.axhline(y = -1.65, linewidth=1.2, linestyle = "--", color = "red")
    g.axvline(x = np.median(X[v]), linewidth=1.2, linestyle = "--", color = "black")
    g.set(xlabel=v, ylabel=target)
```

Các hiệp biến FEV1, FVC và VA có tương quan mạnh với TLC.

## Dựng mô hình DML

Phương pháp DML có cơ chế như sau:

Chúng ta cần 3 mô hình, tạm đặt tên là MT, MY và Metalearner

MT là mô hình phân loại có mục tiêu tiên đoán bệnh nhân có được điều trị bằng 2 loại thuốc nói trên hay không ? Đây là 1 bài toán phân loại nhiều nhãn (MultiLabel) (chú ý: không phải Multiclass, vì 1 bệnh nhân có thể được chỉ định dùng cùng lúc cả 2 loại thuốc). Để tạo ra MT, ta dùng mô hình Extreme Gradient boosting (XGB_Classifier) phân loại nhị phân, lồng trong một mô hình MultiOutputClassifier của scikitlearn. Dữ liệu đầu vào của MT gồm W và X blocks

MY là mô hình hồi quy có mục tiêu ước tính càng chính xác càng tốt giá trị của outcome từ X và W blocks (chú ý dữ liệu đầu vào của MY không bao gồm các biến Treatment T1 và T2). Ta dùng 1 mô hình XGB regressor cho MY

Lúc này, ta quan tâm đến Residual error của 2 mô hình MT và MY, thí dụ Residual của MY là phần variance của Outcome mà các biến X,W không đủ khả năng giải thích được; Nếu ta dùng 1 mô hình thứ 3 : Meta learner, có khả năng ước lượng residual MY theo residual MT, ta đã vô hiệu hóa hiệu ứng nhiễu của tất cả counfouders và covariates, một mô hình như vậy sẽ cho phép ước tính đại lượng mà ta gọi là CATE (Conditional average treatment effect) của 2 loại thuốc. CATE có cùng đơn vị và thang đo với Outcome, nó là phần variance của outcome trực tiếp do Treatment chi phối.

```python
xgbc = XGBClassifier(n_estimators=500,
                     grow_policy = "lossguide",
                     eval_metric = 'logloss')

xgbreg_n = XGBRegressor(objective = 'reg:squarederror',
                      n_estimators=500,
                      grow_policy = "lossguide",
                      eval_metric = 'mae')

xgbreg_final = XGBRegressor(n_estimators=500,
                            objective='reg:squarederror',
                            grow_policy = "lossguide",
                            eval_metric = 'mae')
```

Ta huấn luyện hệ thống 3 mô hình này, kết quả là object tên 'est', với tùy chỉnh inference = 'bootstrap', mô hình metalearner cho phép ước tính khoảng tin cậy của CATE1 và CATE2:

```python
# Estimator Z-score

est = DMLCateEstimator(model_y = xgbreg_n,
                model_t = MultiOutputClassifier(xgbc),
                model_final = xgbreg_final)

est.fit(Y = Y.values,
        T = T.values,
        X = X.values,
        W = W.values,
        inference = "bootstrap")
```

## Kết quả

Mô hình DML (est) cho phép ước tính hiệu quả can thiệp CATE cho riêng từng loại thuốc, trên một bệnh nhân bất kì.
Nếu ta áp dụng mô hình này cho toàn bộ bệnh nhân trong dữ liệu, ta sẽ có CATE1 và CATE2 của mẫu khảo sát:

```python
marg_ef = est.marginal_effect(X = X.values, T = T.values)

ml, mu = est.marginal_effect_interval(X = X.values, T=T.values)
```

Đối với Vasodilator (thuốc dãn mạch), CATE được ước tính cho sample có median = -0.16; 32.4% mật độ phân phối là >0 (hiệu quả tốt), 67.6% có CATE <0 (hiệu quả không tốt)

```python
az.plot_posterior(marg_ef[:,:,0], 
                  credible_interval = 0.95,
                  point_estimate = 'median',
                  ref_val = 0,
                  **{'alpha': 0.8, 'color':'red'})
```

Đối với Immunosupressant (thuốc ức chế miễn dịch), CATE được ước tính cho sample có median = 0.17 (khá tốt); 65.7% mật độ phân phối là >0 (hiệu quả tốt), 34.3 % có CATE <0 (hiệu quả không tốt)

```python
az.plot_posterior(marg_ef[:,:,1], 
                  credible_interval = 0.95,
                  point_estimate = 'median',
                  ref_val = 0,
                  **{'alpha': 0.8, 'color':'blue'})
```

Tuy nhiên, theo giả định thì việc ước tính trung vị của CATE cho cả sample không thực sự nói lên điều gì cả, vì thời gian mắc bệnh là 1 covariate quan trọng, TLC chỉ có thể giảm chứ không bao giờ tăng được, như ta nói. Do đó cách diễn giải có ý nghĩa hơn là khảo sát diễn tiến của CATE theo Duration:

```python
size = X.shape[0]

marginal_eff_df = pd.DataFrame({'Duration': X['Duration'].values,
                                'Vasodilat_CATE': marg_ef[:,:,0].reshape(size),
                                'Immunosup_CATE': marg_ef[:,:,1].reshape(size),
                                'Vasodilat_LB': ml[:,:,0].reshape(size),
                                'Immunosup_LB': ml[:,:,1].reshape(size),
                                'Vasodilat_UB': mu[:,:,0].reshape(size),
                                'Immunosup_UB': mu[:,:,1].reshape(size),})
```

```python
def CATE_plot(effect = 'Vasodilat', line_col = "darkred"):
    plt.figure(figsize=(8,6))
    
    sns.regplot(data = marginal_eff_df,
            x = 'Duration',
            y =  effect + '_CATE',
            lowess = True, marker = ".",
            color = line_col,
            ci = 95,
            scatter_kws={'alpha':0.3, 'color':'grey'})
    
    sns.regplot(data = marginal_eff_df,
            x = 'Duration',
            y = effect + '_LB',
            lowess = True, marker = ".",
            color = "grey",
            ci = 95,
            scatter_kws={'alpha':0, 'color':None})
    
    sns.regplot(data = marginal_eff_df,
            x = 'Duration',
            y = effect  + '_UB',
            lowess = True, marker = ".",
            color = "grey",
            ci = 95,
            scatter_kws={'alpha':0, 'color':None},)
    
    plt.axhline(linewidth=1.5, linestyle = "--", color = "red")
    
    plt.xlabel('Duration (years)')
    plt.ylabel('CATE_Vasodilator')
```

```python
CATE_plot(effect = 'Vasodilat', line_col = "darkred")
```

```python
CATE_plot(effect = 'Immunosup', line_col = "blue")
```

Theo kết quả này, hiệu quả CATE trung bình của thuốc ức chế miễn dịch là tốt trong khoảng thời gian ngay sau khi chẩn đoán bệnh và ổn định trong khoảng 10 năm, sau đó giảm dần trong 30 năm sau. Tuy nhiên hiệu quả này tốt hơn so với thuốc dãn mạch.
